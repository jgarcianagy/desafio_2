#### AMBIENTES

Una de las columnas que tiene bastantes nulos es rooms.
Vamos a intentar extraer información sobre la cantidad de ambientes de las 
columnas descripcion y titulo. 


```python
#busqueda de info de los ambientes en la descripcion

patron_un_amb = 'un|uno'
patron_dos_amb = 'dos'
patron_tres_amb = 'tres'
patron_cuatro_amb = 'cuatro'
regex_1amb = re.compile(patron_un_amb, flags = re.IGNORECASE)
regex_2amb = re.compile(patron_dos_amb, flags = re.IGNORECASE)
regex_3amb = re.compile(patron_tres_amb, flags = re.IGNORECASE)
regex_4amb = re.compile(patron_cuatro_amb, flags = re.IGNORECASE)
#subtituimos los numeros en letras por dígitos
#Description
serie_ambientes_desc_dig = data_selected.description.apply(lambda x: x if x is np.NaN else regex_1amb.sub("1",x) )
serie_ambientes_desc_dig = serie_ambientes_desc_dig.apply(lambda x: x if x is np.NaN else regex_2amb.sub("2",x) )
serie_ambientes_desc_dig = serie_ambientes_desc_dig.apply(lambda x: x if x is np.NaN else regex_3amb.sub("3",x) )
serie_ambientes_desc_dig = serie_ambientes_desc_dig.apply(lambda x: x if x is np.NaN else regex_4amb.sub("4",x) )
#Title
serie_ambientes_title_dig = data_selected.title.apply(lambda x: x if x is np.NaN else regex_1amb.sub("1",x) )
serie_ambientes_title_dig = serie_ambientes_title_dig.apply(lambda x: x if x is np.NaN else regex_2amb.sub("2",x) )
serie_ambientes_title_dig = serie_ambientes_title_dig.apply(lambda x: x if x is np.NaN else regex_3amb.sub("3",x) )
serie_ambientes_title_dig = serie_ambientes_title_dig.apply(lambda x: x if x is np.NaN else regex_4amb.sub("4",x) )
```

```python
#Buscamos los digitos que esten al lado de palabras que empiecen con dorm o amb
patron_amb = "(\d\sdorm)|(\d\samb)"
regex_amb = re.compile(patron_amb, flags = re.IGNORECASE)

#Description
serie_amb_desc = serie_ambientes_desc_dig.apply(lambda x: x if x is np.NaN else regex_amb.search(x) )
#nos quedamos solo con los digitos para armar una serie solo con los numeros
patron_dig = '\d\s'
regdig = re.compile(patron_dig)
nro_ambientes_desc = serie_amb_desc.apply(lambda x: x if x is np.NaN else regdig.search(str(x)))

#Title
#busca los digitos de la serie extraida de title
serie_title_amb = serie_ambientes_title_dig.apply(lambda x: x if x is np.NaN else regex_amb.search(x) )
# de la serie anterior, que tenia "2 amb" o "3 dormitorios" nos quedamos solo con el numero
nro_ambientes_title = serie_title_amb.apply(lambda x: x if x is np.NaN else regdig.search(str(x)))
```

```python
pd.options.mode.chained_assignment = None # para que no aparezca el warning de pandas

#Armando las series (los armé en columnas porque no sabía bien como hacer para que me quede en series tomando la mascara bien)
#Description
mask_nro_ambientes_not_null = nro_ambientes_desc.notnull()
data_selected.loc[mask_nro_ambientes_not_null, "ambientes_desc"] = nro_ambientes_desc[mask_nro_ambientes_not_null].apply(lambda x: x.group(0)).astype(int)

mask_nro_ambientes_title_not_null = nro_ambientes_title.notnull()
data_selected.loc[mask_nro_ambientes_not_null, "ambientes_title"] = nro_ambientes_title[mask_nro_ambientes_title_not_null].apply(lambda x: x.group(0)).astype(int)
```

```python
#combinamos la columa de ambientes extraidos en description con los extraidos de title
ambientes_combinados = data_selected.ambientes_desc.combine(data_selected.ambientes_title, max, fill_value=0)
#combinamos el resultado con las que ya estaban en rooms
ambientes_combinados2 = ambientes_combinados.combine(data_selected.rooms, max, fill_value = 0)
#comparamos combinar o no lo extraido de la descripcion y del titulo con combinarlo tambien con room, a ver si suma o no
print("Nulos:")
print("Combinado sin rooms: ", ambientes_combinados.isnull().sum()," Combinado con rooms ", ambientes_combinados2.isnull().sum(), " Rooms solo:" , data_selected.rooms.isnull().sum())
```

```python
#creamos la columna ambientes
data_selected['ambientes'] = ambientes_combinados
data_selected.drop(columns= ['ambientes_desc','ambientes_title'], inplace = True)
data_selected.columns
```

```python
(data_selected.ambientes.isnull().sum() / len(data_selected) * 100).round(2)
```

#### AMENITIES

De la descripción y el título vamos a extraer información de las características o amenities de las propiedades y armar columnas con la información separada. 

```python
 #Esta funcion es para pasarle un patron de regex y devuelva una columna con 0 y 1 si está el elemento en el titulo o en la descripcion
def armar_serie_con_regex(patron):
    #aplico el patron y lo busco en la columna description por un lado y en title por otro
    regex_gym = re.compile(patron, flags = re.IGNORECASE)
    serie_a_armar_desc = data_selected.description.apply(lambda x: x if x is np.NaN else regex_gym.search(x) )
    serie_a_armar_title = data_selected.title.apply(lambda x: x if x is np.NaN else regex_gym.search(x) )
    #a los datos positivos extraidos de description les asigno un 1 y a los nulos un 0
    mascara_serie_desc_notnull = serie_a_armar_desc.notnull()
    serie_a_armar_desc[mascara_serie_desc_notnull] = 1 
    serie_a_armar_desc.fillna(0, inplace = True)
    #a los datos positivos extraidos de title les asigno un 1 y a los nulos un 0
    mascara_a_armar_title_notnull = serie_a_armar_title.notnull()
    serie_a_armar_title[mascara_a_armar_title_notnull] = 1 
    serie_a_armar_title.fillna(0, inplace = True)
    #combino ambas series 
    serie_armada = serie_a_armar_desc.combine(serie_a_armar_title, max, fill_value=0).astype(int)
    return serie_armada
```

```python
# Para probar el modelo, tomamos los datos de CABA, Barrio Recoleta, columnas 'price_aprox_usd' y 'surface_covered_in_m2'
mask_caba = data_selected.state_name == 'Capital Federal'
data1 = data_selected[mask_caba]

mask_recoleta = data1.place_name == 'Recoleta'
data2 = data1[mask_recoleta]
data3 = data2[['price_aprox_usd','surface_covered_in_m2']]
data4 = data3.dropna()
data4.isnull().sum()

# Cambiamos el nombre de las columnas por 'X' y 'Y'
data5= data4.rename(columns={'price_aprox_usd':'y','surface_covered_in_m2':'x'})

# Convertimos las series en Arrays para que corra en el sklearn
# https://pandas.pydata.org/pandas-docs/version/0.24.0rc1/api/generated/pandas.Series.to_numpy.html

x_array = data5.x.to_numpy()
# print(x_array.astype(int))

y_array = data5.y.to_numpy()
# print(y_array.astype(int))
```

```python
data1.to_csv("data_desafio2.csv", index = False)
```

```python
# Importamos la biblioteca para tirar la regresion y obtener intercepto y curva
# regresion. ver = https://realpython.com/linear-regression-in-python/

import numpy as np
from sklearn.linear_model import LinearRegression

x = x_array.reshape((-1, 1))
y = y_array


model = LinearRegression()

model.fit(x, y)

model = LinearRegression().fit(x, y)

# Obtenemos los elementos de la función predictiva:

r_sq = model.score(x, y)

print('Los elementos de mi función son:')

print('Coeficiente de determinación R2:', r_sq)

print('Intercepto:', str(int(model.intercept_)))

print('Curva:', str(int(model.coef_)))

print('')


# Predecimos asignando la cantidad de metros de nuestro depto en Recoleta a la variable XP:
# Suponemos que el depto tiene 55 metros

xp = np.array([55]).reshape((-1, 1))

y_pred = model.predict(xp)

print("El valor de nuestro departamento de 55mts puede ser " + str(int(y_pred[0])))
```
